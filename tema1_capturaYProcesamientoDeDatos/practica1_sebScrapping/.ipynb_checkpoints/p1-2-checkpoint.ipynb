## Nulos 

Los nulos o huecos, también conocidos por el término en inglés `missing values` aparecen en muchos conjuntos de datos y tratarlos correctamente es importante. 

Un valor nulo, que se representa en python como None o NaN, puede significar varias cosas

- Un valor desconocido; por ejemplo en una columna edad puede que no sepamos la edad de una persona
- Un valor que no tiene sentido para esa fila; por ejemplo ejemplo podemos tener datos de móviles que incluyen su capacidad gráfica y tener móviles antiguos que trabajan en texto y dejan esta columna vacía porque no tiene sentido
- Un error; por ejemplo al calcular porcentajes hemos hecho una división por 0

Hay que conocer de qué tipo son nuestros "huecos" porque muchos métodos de machine learning no funcionarán (o peor, tendrán un comportamiento extraño) en presencia de nulos por lo que a menudo no podremos simplemente ignorarlos.

```python
import pandas as pd
df = pd.DataFrame([[1.0,20],[3.0,None],[4.0,25]],columns=["A","edad"])
df
```

```python
# media de la columna A
sum(df.A)/len(df.A)
```

```python
# media de la columna edad
sum(df.edad)/len(df.edad)
```

Por fortuna en ocasiones Pandas será capaz de resolver el problema. Por ejemplo en el caso de la media podemos escribir:


```python
df.edad.mean()
```

Como vemos ha ignorado los nulos tanto para la suma de edades como para el recuento (dividiendo correctamente entre 2). Puede de todas formas que esto nos interese y prefiramos que nos dé un error si hay nulos. Mirando la [ayuda](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mean.html) de `mean` vemos que podemos decirle que tenga en cuenta los nulos

```python
df.edad.mean(skipna=False)
```

```python

import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/RafaelCaballero/tdm/master/datos/NATOships.csv")
df
```

Una pregunta que debemos hacernos es sí siempre que length es nula lo es beam y viceversa; podrían ser 34 en filas distintas para cada columna. Para esto es muy útil la librería missingno que nos lo muestra de forma gráfica

```python
import missingno as msno
import matplotlib
import matplotlib.pyplot as plt
msno.matrix(df)
plt.show()
```

**Pregunta 1** Vemos que sí que coinciden; podríamos quizás eliminar las filas con algún nulo o estas dos columnas (con `dropna`); sin embargo nos decidimos por otra posibilidad: queremos rellenar los huecos con la media de cada columna (media sin tener en cuenta los huecos, claro)

Escribir código para generar un nuevo dataframe df2 que sea igual a df salvo que los huecos en las columnas length y mean son reemplazados por la media (pista, buscar información sobre `fillna`)

```python
## solución
reaplaceValue = {'course' : df.course.mean(), 'speed' : df.speed.mean(), 'status' : "-",
                 'flag' : "-", 'length' : df.length.mean(), 'beam' : df.beam.mean(),
                 'lat' : df.lat.mean(), 'long' : df.long.mean(), 'MMSI' : "-"}
df2 = df.fillna(reaplaceValue)
```

```python
# para probar
msno.matrix(df2)
plt.show()
```

#### Visualización de coordenadas geográficas

En este apartado vamos a representar los barcos en el mapa con la biblioteca folium a partir de su longitud y latitud. El mapa que obtendremos será de este estilo:

<img src="https://github.com/RafaelCaballero/tdm/blob/master/images/mapaNATOships.png?raw=true"></img>

Características: 
- cada barco será representado por un icono que será o bien "ship" si su speed>0 o "anchor" si speed==0. 
- Además los barcos de flag "United States (USA)" se mostrarán en rojo y el resto en azul. 
- Al hacer click nos mostrará el MMSI y el país

```python
import folium

m = folium.Map(zoom_start=1.5,location=[60,-60])
for i,r in df2.iterrows():
    folium.Marker(
        location=[float(r["lat"]), float(r["long"])],
        popup="{0}- {1}".format(r["MMSI"],r["flag"]),
        icon=folium.Icon(color="red" if r["flag"]=="United States (USA)" else "blue", icon='ship' if r["speed"]>0 else "anchor", prefix='fa'),
    ).add_to(m)
m
```
