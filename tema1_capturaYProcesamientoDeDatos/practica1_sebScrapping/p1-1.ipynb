Escribir una función `abre`:

- *Entrada*: 
    - driver: variable que contiene la conexión con el navegador (ver código anterior)
    - url: un string con una página web 
- *Descripción*: Abre la página indicada en selenium. Asume que el navegador ya está abierto

Abría la página 

https://www.vesselfinder.com/vessels/NATO-WARSHIP-IMO-0-MMSI-211927000

```python
# solución
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By

def abre(driver,url):
    driver.get(url)
    timeout = 5
    try:
        element_present = EC.presence_of_element_located((By.ID, 'google_esf'))
        WebDriverWait(driver, timeout).until(element_present)
    except TimeoutException:
        print("Timed out, la página no se puede cargar")
    
```

**Ejercicio 1** Escribir una función `coords_s` que obtenga la longitud y la latitud a partir de un texto como el que hay al principio de la página

*Entrada*: cadena de caracteres con el texto al principio de la página

*Salida*: tupla (latitud,longitud) o None si es un string que no tiene el formato adecuado

*Notas*: tener en cuenta que

- latitud: valor real entre -90 (polo sur) y +90 (polo norte)
- longitud: valor real entre -180 y +180. Si da con Este y Oeste, Oeste es negativo, Este es positivo, así 5.065 West sería -5.065

Como el string puede tener el error podemos recoger las excepciones como se indica [aquí](https://stackoverflow.com/questions/4990718/how-can-i-write-a-try-except-block-that-catches-all-exceptions) y en ese caso devolver `None`




```python
# solución

def coords_s(s):
    separator = "."
    
    tempArray = s.split("coordinates ")
    tempArray = tempArray[1].split(" N ")
    if(len(tempArray)<2):
        tempArray = tempArray[0].split(" S ")
    coorN = tempArray[0]
    
    tempArray = s.split(" / ")
    tempArray = tempArray[1].split(" E)")
    if(len(tempArray)<2):
        tempArray = tempArray[0].split(" W)")
    coorE = tempArray[0]
    
    coorPartN = coorN.partition(separator)
    coorPartE = coorN.partition(separator)
    if coorPartN[0].isdigit() and coorPartN[2].isdigit() and (coorPartE[0].isdigit() and coorPartE[2].isdigit()):
        result = (coorN, coorE)
        return result
    else:
        return None
```

**Ejercicio 2** Función `coords`

- *Entrada*:  variable driver de una página ya abierta con datos de un barco 
- *Descripción*: lo único que tiene que hacer es buscar el texto donde están la longitud y la latitudl al principio de la página, extraer dicho y pasárselo a la función anterior `coords_s`, devolviendo lo que devuelva ésta. 
- *Salida*: devuelve su latitud y longitud

Nota: Como puede ser que la página esté mal o no exista el elemento que buscamos podemos añadir excepciones como NoSuchElementException (ver un ejemplo de uso [aquí](https://stackoverflow.com/questions/38022658/selenium-python-handling-no-such-element-exception) ). En este caso se devolverá None.

```python
from selenium.common.exceptions import NoSuchElementException

def coords(driver):

    # Existen dos elementos con la clase text2, pero sabemos que el texto buscado siempre será el primero 
    # debido a la posición del msimo sobre el HTML
    elements=driver.find_elements(By.CLASS_NAME, "text2")
    if len(elements)>=1:   
        result = coords_s(elements[0].text)
        if result != None:  
            return coords_s(elements[0].text)
        else:
            return None
    else:
        return None
```

**Ejercicio 3** Queremos extraer los siguientes valores de la tabla:

- course: número real con el curso (un número entre 0 y 360 sin incluir este último). Es el primer elemento de la casilla `Course / Speed`. tener cuidado de quitar el valor ° antes de convertir a número real
- speed: número real con velocidad en nudos, segundo valor en la casilla  `Course / Speed`.
- status: un string, corresponde a la casilla `Navigation Status`
- flag: string con bandera (país), corresponde a la casilla `Flag`
- length: número real con el "largo" del barco. Primer valor de la casilla `Length / Beam`.
- beam: número real con el "ancho" del barco. Segundo valor de la casilla `Length / Beam`.

Escribir una función `tabla` que reciba una variable driver con la página abierta y devuelva una lista con estos valores (si falta alguno se pondrá el valor None, si no hay tabla se devolverá la lista vacía)

```python
def tabla(driver):
    
    elements=driver.find_elements(By.XPATH, "html//td")
    for index, el in enumerate(elements):
        
        if el.text=='Course / Speed':
            tempArray =elements[index+1].text.split("° / ")
            if(len(tempArray)>1):
                course = tempArray[0]
                tempArray =tempArray[1].split(" kn")
                speed = tempArray[0]
            else:
                course = "-"
                speed = "-"
            
        elif el.text=='Navigation Status':
            status =elements[index+1].text
            
        elif el.text=='Flag':
            flag =elements[index+1].text
            
        elif el.text=='Length / Beam':
            tempArray =elements[index+1].text.split(" / ")
            if(len(tempArray)>1):
                length = tempArray[0]
                tempArray =tempArray[1].split(" m")
                beam = tempArray[0]
            else:
                length = "-"
                beam = "-"
        
    result=(course, speed, status, flag, length, beam)    
    return result
 
```

**Ejercicio 4** Escribir una función datos_barco que unifique todo lo anterior

- *Entrada*: 
    - driver: variable que contiene la conexión con el navegador (ver código anterior)
    - url: un string con una página web 
- *Descripción*: Abre la página indicada en selenium. Asume que el navegador ya está abierto. Obtiene los datos del barco y los devuelve
- *Salida*: lista de la forma  [course,speed,status,flag,length,beam,lat,long] o None si hay un error al extraer la información

```python
def datos_barco(driver,url):
    abre(driver,url)
    coordResult = coords(driver)
    if coordResult != None:
        result = coordResult + tabla(driver)
        return result
    else:
        return None 
```

**Ejercicio 5** Escribir una función `muchos_barcos` con la siguientes características

- *Entrada*: Nombre de un fichero csv que tiene una columna MMSI con identificadores de barcos

- *Descripción*: Tras leer el fichero en panda solo hay que, para valor x de la columna MMSI, 
    - pegar el valor x al final de "https://www.vesselfinder.com/vessels/NATO-WARSHIP-IMO-0-MMSI-"; eso nos dará la URL. Nota: es posible que x sea un entero, en ese caso se puede convertir a str con str(x)
    - utilizar datos_barco para abrir la página correspondiente y obtener los datos del barco

- *Salida*: Un dataframe con los datos de los barcos leídos. Las columnas serán "course","speed","status","flag","length","beam","lat","long","MMSI" (notar que se añade el identificador MMSI al final)

```python
import os
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager


def muchos_barcos(nombre):
    
    chromedriver = "/usr/local/share/chromedriver"
    os.environ["webdriver.chrome.driver"] = chromedriver
    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--no-sandbox')
    driver = webdriver.Chrome( service=Service(ChromeDriverManager().install()), options=chrome_options)
    
    boardData_course =[]
    boardData_speed =[]
    boardData_status =[]
    boardData_flag =[]
    boardData_length =[]
    boardData_beam =[]
    boardData_lat =[]
    boardData_long =[]
    boardData_MMSI =[]
    
    fileMMI = pd.read_csv(nombre)  
    for index in range(len(fileMMI)):
        url = "https://www.vesselfinder.com/vessels/NATO-WARSHIP-IMO-0-MMSI-{url}".format(url=fileMMI.loc[index][0])
        boardDataTemp = datos_barco(driver,url)
        if boardDataTemp != None:
            boardData_course.append(boardDataTemp[0])
            boardData_speed.append(boardDataTemp[1])
            boardData_status.append(boardDataTemp[2])
            boardData_flag.append(boardDataTemp[3])
            boardData_length.append(boardDataTemp[4])
            boardData_beam.append(boardDataTemp[5])
            boardData_lat.append(boardDataTemp[6])
            boardData_long.append(boardDataTemp[7])
            boardData_MMSI.append(fileMMI.loc[index][0])
            
        if (index%5) == 0: # El objetivo de esta sentencia es simplemente mostrar el avance de la lista con el objetivo de informar al usuario
            print(index,"/",len(fileMMI),"   ", sep="", end="")
                
    result = pd.DataFrame({'course' : boardData_course, 'speed' : boardData_speed, 'status' : boardData_status,
                            'flag' : boardData_flag, 'length' : boardData_length, 'beam' : boardData_beam,
                            'lat' : boardData_lat, 'long' : boardData_long, 'MMSI' : boardData_MMSI})
    driver.close()
    return result
```
